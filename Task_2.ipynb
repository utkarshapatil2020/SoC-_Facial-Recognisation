#Stats

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df=pd.read_csv('/content/Dataset.csv')


df=df.head()


print('mean :',
      '\n',
    df.mean(),
    '\n',
    
    'skew :','\n',
    df.skew(),
    '\n',

    'min :','\n',
    df.min(),
    '\n',
    'max :','\n',
    df.max(),
    '\n',

    'median :','\n',
    df.median(),
    '\n',

    'variance :','\n',
    np.std(df),
    '\n',
    '75 percentile :','\n',
    df.quantile(.75, axis = 0),
    '\n',
    '90 percentile :','\n',
    df.quantile(.90,0),
    
    '\n',
    'inter quantile range :','\n',
    df.quantile(0.5,0)

  

  


)
```
